## Create new data set with select columns

```python
food2011 = food_df[["Area Abbreviation","Area","Item Code", "Item","Element Code","latitude","longitude","Y2011","Unit"]]
food2011.head()
```

```python
###filtering by type of production (food)
food_2011 = food2011.loc[(food2011["Element Code"] == 5142)]
food_2011.head()
```

```python
### let's group by country and sum the total production of food per country
Food_Total = food_2011.groupby(['Area']).sum()['Y2011']
Foods = pd.DataFrame({'Production Totals': Food_Total,})

Foods.head()
```

```python
###let's create the data frame by merging both tables
Food_data = pd.merge(food_2011, Foods, how ="left", on=["Area", "Area"])
Food_data.head()
```

```python
###The precipitation data is coming from the link below: Average precipitation in depth (mm per year). 
###Average precipitation is the long-term average in depth (over space and time) of annual precipitation in the country. 
###Precipitation is defined as any kind of water that falls from clouds as a liquid or a solid.

url = 'https://www.nationmaster.com/country-info/stats/Geography/Average-precipitation-in-depth/Mm-per-year#2011'
```

```python
### creating tables from webpage

tables = pd.read_html(url)
tables
```

```python
### creating the table

df = tables[0]
df.columns = ['ID','COUNTRY', 'RAIN (mm)', 'DATE', 'blank', 'blank 2']
df
```

```python
### dropping columns

DF_2 = df.drop(['ID', 'blank', 'blank 2', 'DATE'], axis=1)
DF_2
```

```python
###renaming COUNTRY

DF_2 = DF_2.rename(columns = {'COUNTRY': 'Area'})
DF_2
```

```python
###let's create the data frame FOOD_DATA + DF_2 
Food_Rain = pd.merge(Food_data, DF_2, how ="left", on=["Area", "Area"])
Food_Rain.head()
```

```python
### Now, we want to incorporate temperature data to this analysis. This csv file has the annual temp for the year 2011.
### https://datacatalog.worldbank.org/dataset/climate-change-knowledge-portal-historical-data


Temperature_data = 'C:/Users/CAEAIZPU/OneDrive - kochind.com/Desktop/ETL Project/cckp_historical_data_0.csv'
```

```python
### reding the file

Temperature_df = pd.read_csv(Temperature_data)
Temperature_df.head()
```

```python
Temperature_df.columns
```

```python
Temp_1=Temperature_df.drop(['Jan_Temp', 'Feb_temp', 'Mar_temp', 'Apr_Temp',
       'May_temp', 'Jun_Temp', 'July_Temp', 'Aug_Temp', 'Sept_temp',
       'Oct_temp', 'Nov_Temp', 'Dec_temp'], axis=1)
Temp_1.head()
```

```python
###lets rename Area abbreviation

Temp_2 = Temp_1.rename(columns = {'ISO_3DIGIT': 'Area Abbreviation'})
Temp_2.head()
```

```python
###let's create the data frame Food_Rain + Temp_2 
Food_Temp = pd.merge(Food_Rain, Temp_2, how ="left", on=["Area Abbreviation", "Area Abbreviation"])
Food_Temp.head()
```

```python
import requests as req
```

```python
### from weather API i want to collect sunrise and sunset to compare daytime per country

# Save config info
url = "http://api.openweathermap.org/data/2.5/weather"

params = {"appid": "fb5402f7576ca409e1bbcb49d0ba91d9",
          "units": "metric"}

# Use 'for' loop to retrieve weather info by rows in dataframe
for column, row in df.iterrows():
    
    # Reference rows for 'q' param
    params["q"] = f'{row["COUNTRY"]}'
    
    # Get and print links for each country
    print(f'Weather information for {params["q"]}')
    weather_response = req.get(url, params)
    print(weather_response.url)
    weather_response  = weather_response.json()
    
    # Get weather data and input into dataframe
    df.set_value(column, "Max Temperature", weather_response.get("main", {}).get("temp_max"))
    df.set_value(column, "Min Temperature", weather_response.get("main", {}).get("temp_min"))
    df.set_value(column, "Sunrise", weather_response.get("sys", {}).get("sunrise"))
    df.set_value(column, "Sunset", weather_response.get("sys", {}).get("sunset"))
```

```python
df.head()
```

```python
new_df = df.drop(['ID', 'DATE', 'blank', 'Latitude', 'Longitude', 'blank 2', 'RAIN (mm)', 'Max Temperature','Min Temperature'], axis=1)
```
